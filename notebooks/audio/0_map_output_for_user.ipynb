```kotlin
import kotlinx.serialization.Serializable
import kotlinx.serialization.json.Json

@Serializable
data class Segment(
    val filename: String,
    val start: Double,
    val end: Double
)

val json = Json {
    prettyPrint = true
}
```

```kotlin
import java.io.File

val output = File("./K_zrdqPHKbk_laughter_detection_edited.json")
val string = output.readText()
val data = json.decodeFromString<List<Segment>>(string)
val dataForUser = data.fold("") {acc, segment ->
    val youtubelink = "https://www.youtube.com/watch?v=K_zrdqPHKbk&t=${segment.start}s"
    acc + youtubelink + "\n"
    acc + ""
}
dataForUser
```
